```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as pt
```

```python
data = pd.read_csv('Dataset//Videogames Data.csv', encoding='mac_roman')
```

```python
data.head()
```

```python
data.fillna('', inplace=True)
```

```python
# data.fillna('', inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.drop('Number of Meta Reviews', 1)
data = data.drop('Number of User Reviews', 1)
data = data.drop('Critic Positive', 1)
data = data.drop('Critic Mixed', 1)
data = data.drop('User Positive', 1)
data = data.drop('User Mixed', 1)
data = data.drop('User Negative', 1)
data = data.drop('Critic Negative', 1)
```

```python
data.isnull().sum()
```

```python
data['User Score'] = pd.to_numeric(data['User Score'])
```

```python
data.info()
```

```python
data = data.drop('User Score', 1)
```

```python
data.drop_duplicates(subset=['Title'], inplace=True)
```

```python
data.head()
```

```python
def remove_spaces(row):
    x=[]
    for i in row:
        i=i.replace(' ','')
        i=i.lower()
        x.append(i)
    return(x)

data['Genre'] = data['Genre'].apply(remove_spaces).str.join('')
data['Description'] = data['Description'].str.lower()
data['ESRB Rating'] = data['ESRB Rating'].apply(remove_spaces).str.join(' ')

data.head()
```

```python
import rake_nltk
from rake_nltk import Rake
r=Rake()
def extract_important_words(row):
    word = r.extract_keywords_from_text(row)
    word = r.get_ranked_phrases()
    return ' '.join(word)

data['Description'] = data['Description'].apply(extract_important_words)
data.head()
```

```python
data['combined_features'] = data['Genre']+' '+data['Description']
data.head()
```

```python
data.drop(columns=['ID','Genre','Description'],inplace=True)
data.head()
```
